# Make Deep Copy of Dataset

```python
df = orders.copy()
df
```


# Added new column and Convert Date Time Format

```python
month = df.Order_date.dt.month
df['month']= df.Order_date.dt.month
df
```

# Avg Sale month on month 

```python
a= df.groupby(by='month')[['Sale_price']].mean()
a.rename(columns={'Sale_price':'Avg_Sale_price'}, inplace=True)
a
```

# No. of orders month on month

```python


no_of_orders= df.groupby(by='month')[['Order_id']].count()
no_of_orders.rename(columns={'Order_id': "no_of_orders"}, inplace= True)
no_of_orders
```

# Total orders month on month by gender

```python
gender_orders= df.groupby(by=['month','Gender'])[['Order_id']].count() 
gender_orders.rename(columns={'Order_id':'No_of_orders'}, inplace = True)
gender_orders
```

# OR

```python
d= df.pivot_table(columns=['Gender'], index=['month'], values=['Order_id'], aggfunc='count', margins=True, margins_name='Total_Orders')
d
```

# Avg Sale x Product category month on month

```python
avg_sale= df.pivot_table(index=['month'], columns=['Product_category'], values=['Sale_price'], margins=True, margins_name='Total_Avg_sale')
avg_sale.rename(columns={'Sale_price': 'Avg_sale_price'}, inplace=True)
avg_sale      
```

# Product Category X Total sale over the period of time and also month on month

```python
total_sale = df.pivot_table(index=['month'], columns=['Product_category'], values=['Sale_price'], aggfunc='sum', margins=True, margins_name='Grand_Total_Sale')
total_sale.rename(columns={'Sale_price':'total_sale'}, inplace=True)
total_sale
```

# Product Category X Order Rating month on month and also over the period of time

```python
rating=df.pivot_table(columns=['Product_category'], index=['Order_rating'], values=['User_id'], aggfunc='count', margins=True, margins_name='Total_Rating')
rating.rename(columns={'User_id':'Rating'})
```

#  Calculate order status month on month

```python
Orders = df.pivot_table(columns=['month'], index=['Order_status'], values=['User_id'], aggfunc='count', margins=True, margins_name='Total_orders')
Orders.rename(columns={'User_id':'Order_status'}, inplace=True)
Orders
```

# No of orders using Offer Name X Product Category

```python
Offer = df.pivot_table(columns = ['Offer_name'], index= ['Product_category'], values=['User_id'], aggfunc='count',margins=True, margins_name='Total')
Offer.rename(columns={'User_id':'No_of_orders'}, inplace=True)
Offer
```
